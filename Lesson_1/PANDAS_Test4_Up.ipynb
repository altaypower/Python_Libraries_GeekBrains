```
import pandas as pd
```

Создайте датафрейм authors_stat на основе информации из authors_price. В датафрейме authors_stat должны быть четыре столбца:
author_name, min_price, max_price и mean_price,
в которых должны содержаться соответственно имя автора,минимальная, максимальная и средняя цена на книги этого автора.



```
a = {
    "author_id": [1, 2, 3],
    "author_name": ['Тургенев', 'Чехов', 'Островский']
}
a
```

```
authors = pd.DataFrame(a)
authors
```

```
b = {
    "author_id": [1, 1, 1, 2, 2, 3, 3],
    "book_title": ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
     "price": [450, 300, 350, 500, 450, 370, 290]
}
b
```

```
book = pd.DataFrame(b)
book
```

```
authors_price = pd.merge(authors, book, on='author_id', how='outer')
authors_price
```

```
price_max = authors_price.groupby("author_name").agg({"price": "max"})
price_max
```

```
price_min = authors_price.groupby("author_name").agg({"price": "min"})
price_min
```

```
price_mean = authors_price.groupby("author_name").agg({"price": "mean"})
price_mean
```

```
authors_stat = pd.DataFrame({"min_price": price_min.price, "max_price": price_max.price, "mean_price": price_mean.price})
authors_stat
```

```
type(authors_stat.min_price)
```

```
authors_stat = authors_price.groupby('author_name').agg({'price': [('min_price', 'min'), ('max_price', 'max'), ('mean_price', 'mean')]})
authors_stat
```
